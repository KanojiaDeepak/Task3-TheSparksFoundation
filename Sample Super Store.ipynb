# Observation
- The Super Store data is of the country United States.
- The Super Store sell its products in 49 states and 531 cities.
- The top city and state in terms of frequency are 'New York' and 'California' respectively.
- The Super Store uses 4 ship modes to deliver the product. They are 'First Class', 'Second Class', 'Standard Class' and 'Same day'. The 'Standard Class' ship mode is used the most by the customers.
- There are three different segments. 'Consumer', 'Corporate Segment' and 'Home Office'. The 'Consumer' Segment has the highest count.
- The Western region has the most sales and Southern region has the lowest sales.
- The superstore sells their products in 3 categories namely 'Office Supplies', 'Furniture' and 'Technology'.
- Under these categories, the products are further divided into 17 sub-categories. 
- 'Office Supplies' category has the highest count among all the categories and 'Binders' has the highest count among all sub-categories.
- Sales is highly corelated to profit and quantity.
- Total Sales = 2296195.5903
- The Western Region has the most sales i.e. 31.59% of total sales.
- The Southern Region has the lowest sales i.e. 17.06% of total sales.
- The 'Technology' category generate the maximum sales.
- Products under the category of 'Technology' has the highest sales and the preferred Ship Mode is 'Same Day'. 'Second Class' ship mode is the least preferred for products under 'Technology'.
- The 'Copiers' subcategory generate the maximum sales.
- Total Profit = 286241.4226
- The Western Region also contributes the most in total profit i.e.37.85% of total profit.
- The Central Region contributes the least in total profit i.e. 13.85% of total profit.
- The 'Technology' category generate the maximum profit. 
- The 'Copiers' subcategory generate the maximum profit.
- The 'Tables' subcategory generate the maximum loss.
